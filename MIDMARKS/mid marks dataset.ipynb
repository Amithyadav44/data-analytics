```python
import pandas as pd

# Load the Excel file
file_path = "MIDMARKS.xlsx"  # Update with the correct path
df = pd.read_excel(file_path, sheet_name="SEM2 MID 1 - ALPHA")

# Drop non-subject columns (assuming first two columns are not subjects)
subject_columns = df.columns[2:]

# Ensure only numeric columns are used
df[subject_columns] = df[subject_columns].apply(pd.to_numeric, errors='coerce')

# Group by section and calculate total and average marks
total_marks = df.groupby("SECTION")[subject_columns].sum()
average_marks = df.groupby("SECTION")[subject_columns].mean()

# Convert to percentage (assuming full marks is 20 for each subject)
subject_percentage = (average_marks / 20) * 100

# Calculate pass percentage (students scoring 10 or more)
pass_marks = 10
pass_percentage = df.groupby("SECTION")[subject_columns].apply(lambda x: (x >= pass_marks).mean() * 100)

# Predict probability of passing (based on past pass rates)
predicted_probability = pass_percentage / 100

# Display the results in table format
print("Total Marks for Each Subject by Section:")
print(total_marks.to_string())

print("\nAverage Percentage for Each Subject by Section:")
print(subject_percentage.to_string())

print("\nPass Percentage for Each Subject by Section:")
print(pass_percentage.to_string())

print("\nPredicted Probability of Passing for Each Subject by Section:")
print(predicted_probability.to_string())
```
